```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df=pd.read_csv(mouse_drug_data_to_load)
mouse_drug_df.head()
clinical_trial_df=pd.read_csv(clinical_trial_data_to_load)
clinical_trial_df.head()

# Combine the data into a single dataset
combined_data=pd.merge(mouse_drug_df,clinical_trial_df,on="Mouse ID")

# Display the data table for preview
out_data=combined_data.loc[combined_data["Mouse ID"]=="f234"]
out_data
combined_data.head(100)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 


# Convert to DataFrame
sub_data=combined_data[["Drug","Timepoint","Tumor Volume (mm3)"]]
sub_data_df=sub_data.groupby(["Drug","Timepoint"]).mean().reset_index()
# # # Preview DataFrame

sub_data_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame
sem_df=sub_data.groupby(["Drug","Timepoint"]).sem().reset_index() 

# Preview DataFrame
sem_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
dtt_data = pd.pivot_table(sub_data_df, values='Tumor Volume (mm3)', index=['Timepoint'],columns=['Drug'])
# Preview that Reformatting worked
dtt_data
```

```python
# Generate the Plot (with Error Bars)
fig = plt.figure()                                                               
ax = fig.add_subplot(1,1,1) 

#calculation and plotting line and Std Err for Tumor Response To Treatment
####### Drug Capomulin
Capomulin_data=sub_data_df.loc[sub_data_df['Drug']=='Capomulin']
Capomulin_data
Capomulin_ln,=plt.plot(Capomulin_data['Timepoint'],Capomulin_data['Tumor Volume (mm3)'],marker="o",color="red", linewidth=1,linestyle=':', label="Capomulin")

# Std Err for Drug Capomulin
Capomulin_mean=sub_data_df.loc[sub_data_df['Drug']=='Capomulin','Tumor Volume (mm3)']
Capomulin_stderr=sem_df.loc[sem_df['Drug']=='Capomulin','Tumor Volume (mm3)']
err_ln=ax.errorbar(Capomulin_data['Timepoint'], Capomulin_mean, Capomulin_stderr, fmt="o",color="red")

####### Drug Infubinol
Infubinol_data=sub_data_df.loc[sub_data_df['Drug']=='Infubinol']
Infubinol_data
Infubinol_ln,=plt.plot(Infubinol_data['Timepoint'],Infubinol_data['Tumor Volume (mm3)'],marker="^",color="blue", linewidth=1,linestyle=':', label="Infubinol")

# Std Err for Drug Infubinol
Infubinol_mean=sub_data_df.loc[sub_data_df['Drug']=='Infubinol','Tumor Volume (mm3)']
Infubinol_stderr=sem_df.loc[sem_df['Drug']=='Infubinol','Tumor Volume (mm3)']
err_ln=ax.errorbar(Infubinol_data['Timepoint'], Infubinol_mean, Infubinol_stderr, fmt="^",color="blue")

######## Drug 'Ketapril'
Ketapril_data=sub_data_df.loc[sub_data_df['Drug']=='Ketapril']
Ketapril_data
Ketapril_ln,=plt.plot(Ketapril_data['Timepoint'],Ketapril_data['Tumor Volume (mm3)'],marker="s",color="green", linewidth=1,linestyle=':', label="Ketapril")

# Std Err for Drug 'Ketapril'
Ketapril_mean=sub_data_df.loc[sub_data_df['Drug']=='Ketapril','Tumor Volume (mm3)']
Ketapril_stderr=sem_df.loc[sem_df['Drug']=='Ketapril','Tumor Volume (mm3)']
err_ln=ax.errorbar(Ketapril_data['Timepoint'], Ketapril_mean, Ketapril_stderr, fmt="s",color="green")

######### Calculating Placebo related data
Placebo_data=sub_data_df.loc[sub_data_df['Drug']=='Placebo']
Placebo_data
Placebo_ln,=plt.plot(Placebo_data['Timepoint'],Placebo_data['Tumor Volume (mm3)'],marker="D",color="black", linewidth=1,linestyle=':', label="Placebo")

#Std Err for Drug Placebo
Placebo_mean=sub_data_df.loc[sub_data_df['Drug']=='Placebo','Tumor Volume (mm3)']
Placebo_stderr=sem_df.loc[sem_df['Drug']=='Placebo','Tumor Volume (mm3)']
err_ln=ax.errorbar(Placebo_data['Timepoint'], Placebo_mean, Placebo_stderr, fmt="D",color="black")

# Plotting the line Plot (with Error Bars)
#setting the ticks for the plot
major_ticks = np.arange(30, 80, 5)                                                                                                   
ax.set_yticks(major_ticks)  

#Labeling the plot
plt.title("Tumor Response To Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# Legend
plt.legend(handles=[Capomulin_ln,Infubinol_ln,Ketapril_ln,Placebo_ln], loc="best")

#Grid
plt.grid()

# Save the Figure
plt.savefig('./pymaceuticals_Tumor_Response_To_Treatment.png')
```

```python
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
#combined_data
mouse_data=combined_data[["Drug","Timepoint","Mouse ID"]]
mcnt_data_df=mouse_data.groupby(["Drug","Timepoint"]).count().reset_index()
mcnt_data_df=mcnt_data_df.rename(columns={"Mouse ID":"Mouse Count"})
# Preview DataFrame

mcnt_data_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
dtt_mouse_data = pd.pivot_table(mcnt_data_df, values='Mouse Count', index=['Timepoint'],columns=['Drug'])
# Preview the Data Frame
dtt_mouse_data
```

```python
# Generate the Plot (Accounting for percentages)
fig = plt.figure()                                                               
ax = fig.add_subplot(1,1,1)             

# Drug Capomulin
Capomulin_data=mcnt_data_df.loc[mcnt_data_df['Drug']=='Capomulin']
Capomulin_max_count=Capomulin_data["Mouse Count"][0]
Capomulin_data["Survival Percent"]=100*( Capomulin_data["Mouse Count"]/ Capomulin_max_count)
Capomulin_per_ln,=plt.plot(Capomulin_data['Timepoint'],Capomulin_data["Survival Percent"],marker="o",color="red", linewidth=0.5,linestyle=':', label="Capomulin")

# Drug Infubinol
Infubinol_data=mcnt_data_df.loc[mcnt_data_df['Drug']=='Infubinol']
Infubinol_max_count=Infubinol_data["Mouse Count"].max()
Infubinol_data["Survival Percent"]=100*( Infubinol_data["Mouse Count"]/ Infubinol_max_count)
Infubinol_per_ln,=plt.plot(Infubinol_data['Timepoint'],Infubinol_data["Survival Percent"],marker="^",color="blue", linewidth=0.5,linestyle=':', label="Infubinol")


#Drug Ketapril
Ketapril_data=mcnt_data_df.loc[mcnt_data_df['Drug']=='Ketapril']
max_count=Ketapril_data["Mouse Count"].max()
Ketapril_data["Survival Percent"]=100*( Ketapril_data["Mouse Count"]/ max_count)
Ketapril_per_ln,=plt.plot(Ketapril_data['Timepoint'],Ketapril_data["Survival Percent"],marker="s",color="green", linewidth=0.5,linestyle=':', label="Ketapril")

# Placebo
Placebo_data=mcnt_data_df.loc[mcnt_data_df['Drug']=='Placebo']
max_count=Placebo_data["Mouse Count"].max()
Placebo_data["Survival Percent"]=100*( Placebo_data["Mouse Count"]/ max_count)
Placebo_per_ln,=plt.plot(Placebo_data['Timepoint'],Placebo_data["Survival Percent"],marker="D",color="black", linewidth=0.5,linestyle=':', label="Placebo")

#Draw the plot
major_ticks = np.arange(30,110, 10)                                                                                                   
ax.set_yticks(major_ticks)

#Label the plot
plt.title("Survival During treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate(%)")

#Legend
plt.legend(handles=[Capomulin_per_ln,Infubinol_per_ln,Ketapril_per_ln,Placebo_per_ln], loc="best")

#Grid
plt.grid()

# Save the Figure
plt.savefig('./pymaceuticals_SurvivalRate_During_Treatment.png')

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

#Drug Capomulin
Capomulin_max_count=dtt_data["Capomulin"][0]
Capomulin_min_count=dtt_data["Capomulin"][45]

# # # Display the data to confirm
Capomulin_per=100* (Capomulin_min_count-Capomulin_max_count)/Capomulin_max_count
Capomulin_per

#Drug Infubinol
Infubinol_max_count=dtt_data["Infubinol"][0]
Infubinol_min_count=dtt_data["Infubinol"][45]
# # # Display the data to confirm
Infubinol_per=100* (Infubinol_min_count-Infubinol_max_count)/Infubinol_max_count
Infubinol_per

#Drug Ketapril
Ketapril_max_count=dtt_data["Ketapril"][0]
Ketapril_min_count=dtt_data["Ketapril"][45]
# # # Display the data to confirm
Ketapril_per=100* (Ketapril_min_count-Ketapril_max_count)/Ketapril_max_count
Ketapril_per

#Drug Placebo
Placebo_max_count=dtt_data["Placebo"][0]
Placebo_min_count=dtt_data["Placebo"][45]
# # # Display the data to confirm
Placebo_per=100* (Placebo_min_count-Placebo_max_count)/Placebo_max_count
Placebo_per


```

```python
# Store all Relevant Percent Changes into a Tuple
drug_per_ch=[Capomulin_per,Infubinol_per,Ketapril_per,Placebo_per]
drug_per_ch

drug_name=["Capomulin","Infubinol","Ketapril","Placebo"]
drug_name

merged_list = [(drug_name[i], drug_per_ch[i]) for i in range(0, len(drug_name))] 
merged_list

# Splice the data between passing and failing drugs
# neg_list will contain the data for the Drugs causing the reduction in the Total Tumor volume,hence passing drugs
# pos_list will contain the data for the Drugs causing the increase in the Total Tumor volume,hence failing drugs
neg_list=[]
pos_list=[]

# iterating over Tuples list to collect the data for passing and failing drugs
for p in merged_list:
    if p[1]<0:
        neg_list.append(p)
    else:
        pos_list.append(p)
    #print('{} : {}'.format(p[0], p[1]))
#neg_list
#pos_list

#Creating dataframes from the list created above
df_neg_Obj = pd.DataFrame(neg_list)
#df_neg_Obj

df_pos_Obj = pd.DataFrame(pos_list)
#df_pos_Obj

# Orient widths. Add labels, tick marks, etc. 
#Plotting the passing drug bars
bars1=plt.bar(df_neg_Obj[0], df_neg_Obj[1], color='green',width=-1 ,align='edge')

#Plotting the failing drug bars
bars2=plt.bar(df_pos_Obj[0], df_pos_Obj[1], color='r',width=-1 ,align='edge')

#Label the plot
plt.ylabel('% Volume Tumor Change')
plt.title('Tumor Change Over 45 Day Treatment')
plt.ylim(-30,70,20)

#Grid
plt.grid()


# Use functions to label the percentages of changes
def add_lable(rect):
    for bar in rect:
        height = bar.get_height()
        if (height>0):
            plt.text(bar.get_x() + bar.get_width()/2, height - 6, str(height.astype(int))+'%',ha='center', va='bottom',color='white')
        else:
             plt.text(bar.get_x() + bar.get_width()/2, height + 2, str(height.astype(int))+'%',ha='center', va='bottom',color='white')
    
# Call functions to implement the function calls
add_lable(bars1) 
add_lable(bars2) 

# Save the Figure
plt.savefig('./pymaceuticals_TotalTumorPercent_VolumeChange.png')

# Show the Figure
fig.show()
```
